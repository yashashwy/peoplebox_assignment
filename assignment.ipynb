```python
melted_df1=df2.melt(id_vars=["Employee Code"],value_name="Compensation_score")
melted_df2=df3.melt(id_vars=["Employee Code"],value_name="Review Score")
melted_df3=df4.melt(id_vars=["Employee Code"],value_name="Engagement Score")
```
